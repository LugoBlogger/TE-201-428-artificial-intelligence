```python
import numpy as np

from PIL import Image, ImageFilter
```

```python
# Open image
image = Image.open("./john-w-w-bridge.png").convert("RGB")
image
```

```python
# Filter image according to edge detection kernel
filtered = image.filter(ImageFilter.Kernel(
  size=(3, 3),
  kernel=[-1, -1, -1, -1, 8, -1, -1, -1, -1],
  scale=1
))
```
